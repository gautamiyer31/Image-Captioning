```python
# Decreasing size of all the images for Azure
for i in df.index:
    filename = "img_"+str(i)+".png"
    url = df.loc[i].iat[1]
    response = requests.get(url)
    img = Image.open(BytesIO(response.content))
    img.thumbnail((500,500))
    img.save(filename)
```

```python
# English 
desc_en_text = []
desc_en_conf = []
for i in df.index:
    image_path = "img_"+str(i)+".png"

    cred = CognitiveServicesCredentials(key)
    client = ComputerVisionClient(endpoint = endpoint, credentials = cred)

    with open(image_path, 'rb') as img:
        result = client.describe_image_in_stream(img)
    desc_en_text.append(result.captions[0].text)
    desc_en_conf.append(result.captions[0].confidence)
    
    #Image(filename = image_path, width = 120, height = 120)
```
